```python
import dash
from dash import dcc, html
import pandas as pd
import plotly.graph_objs as go

# Carregando dados fictícios
dados_acidentes = pd.DataFrame({
    'Tipo de Acidente': ['Queda', 'Corte', 'Queimadura', 'Lesão Muscular'],
    'Número de Acidentes': [25, 15, 10, 5]
})

# Inicializando o aplicativo Dash
app = dash.Dash(__name__)

# Layout do aplicativo
app.layout = html.Div(children=[
    html.H1(children='Dashboard de Segurança no Trabalho'),

    html.Div(children='''
        Estatísticas de Acidentes de Trabalho.
    '''),

    dcc.Graph(
        id='grafico-barras',
        figure={
            'data': [
                go.Bar(
                    x=dados_acidentes['Tipo de Acidente'],
                    y=dados_acidentes['Número de Acidentes'],
                    marker=dict(color='rgb(55, 83, 109)')
                )
            ],
            'layout': go.Layout(
                title='Número de Acidentes por Tipo',
                xaxis=dict(title='Tipo de Acidente'),
                yaxis=dict(title='Número de Acidentes'),
                margin=dict(l=40, r=40, t=40, b=40)
            )
        }
    ),

    dcc.Graph(
        id='grafico-pizza',
        figure={
            'data': [
                go.Pie(
                    labels=dados_acidentes['Tipo de Acidente'],
                    values=dados_acidentes['Número de Acidentes']
                )
            ],
            'layout': go.Layout(
                title='Distribuição de Acidentes por Tipo',
                margin=dict(l=40, r=40, t=40, b=40)
            )
        }
    )
])

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dados hipotéticos: número de acidentes por departamento por mês
departamentos = ['Departamento A', 'Departamento B', 'Departamento C', 'Departamento D']
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho']
dados_acidentes = np.random.randint(0, 20, size=(len(departamentos), len(meses)))

# Criando um DataFrame
df_acidentes = pd.DataFrame(data=dados_acidentes, index=departamentos, columns=meses)

# Criando o heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_acidentes, annot=True, cmap="YlGnBu", fmt='d')
plt.title('Número de Acidentes de Trabalho por Departamento e Mês')
plt.xlabel('Mês')
plt.ylabel('Departamento')
plt.show()
```
